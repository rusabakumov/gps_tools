```python
from gpstools import viz
from gpstools.parser import *
from gpstools.track.track import *
from gpstools.track.utils import build_sparse_track
from gpstools.track.activity_detection import *
from gpstools.track.normalization import *
from gpstools.ss_analysis.ss_analysis_graph import SSAnalysisGraph
import os
```

```python
PATH_TO_TRACKS='data/vladimir_09_19/'
REFERENCE_TRACK='abakumov_su2.gpx'

SPEED_PARAMS = TrackSpeedParams(use_provided_speed=True, smoothing_1hz=1, smoothing_10hz=1)
ACTIVITY_DETECTION_PARAMS = ActivityDetectionParams(
    align_to_minutes=True,
    speed_eps=2,
    segment_max_allow_pause=1,
    segment_min_duration=100
)
TRACK_NORM_PARAMS = TrackNormalizationParams(align_to_minutes=True, neighbor_weights=None)
```

```python
# Loading and processing reference

# TODO select track type based on file extension
track = load_track(os.path.join(PATH_TO_TRACKS, REFERENCE_TRACK), TRACK_NORM_PARAMS)
track.print_stats()

# viz.generate_tracks_map([track])

ref_segs = get_activity_segments_for_track(track, ACTIVITY_DETECTION_PARAMS)
# for seg in ref_segs:
#     seg.print_stats()

track = track.crop_to_activity_segment(ref_segs[0], 'ref')
# track.print_stats()

#viz.generate_reference_selection_graph([track])

track = track.crop_to_point_idx(2780)
track.print_stats()

reference_track = build_sparse_track(track, distance_threshold=0.002)
reference_track.print_stats()
```
